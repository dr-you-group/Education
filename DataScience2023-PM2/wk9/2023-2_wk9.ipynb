# 웹 크롤링/스크래핑이란?

### 웹 크롤러 (Web crawler)
> *” A Web crawler, sometimes called a spider or spiderbot and often shortened to crawler, is an Internet bot that systematically browses the World Wide Web, typically for the purpose of Web indexing (web spidering). ” (출처: wikipedia)*

웹 인덱싱을 위해 WWW (Wolrd Wide Web) 을 체계적으로 탐색, 색인, 저장하는 도구를 크롤러라고 부르며, 크롤러가 하는 행위를 '크롤링' 이라고 함.

한번 만들어 두면 편리하게 데이터를 수집할 수 있음.

하위 페이지를 차례대로 방문해 돌아다니는 모습이 흡사 거미와 유사하여 스파이더로도 불림.

### 웹 스크래핑 (Web scraping)
> *” Web scraping, web harvesting, or web data extraction is data scraping used for extracting data from websites. ” (출처: wikipedia)*

웹 스크래핑은 다양한 웹사이트로부터 데이터를 추출하는 기술을 의미함.

## 자주 혼동되는 용어 (웹 크롤링 vs. 웹 스크래핑)

정의에 대해 잘 살펴보면, 흔히 사용되고 있는 용어 '웹 크롤링'은 **데이터를 추출하는 것보다, 웹 사이트를 탐색하고 인덱싱 하는 것에 더 주요한 목적**을 가지고 있음.

크롤링: 웹상의 데이터를 추출하거나 색인하기 위해 **체계적으로** 운용됨 / 스크래핑: 원하는 **특정 사이트의 데이터**를 일괄적으로 추출

# HTML 이란?

## HTML은 Hyper-Text-Markup-Language의 약자로, 웹 페이지를 만드는 데 사용하는 언어임. 우리가 브라우저(크롬, 사파리 등)를 통해 보는 페이지는 HTML로 작성되어 있음

## HTML은 마크업 언어로써, 여러 가지 태그를 이용하여 제목이나 본문을 구분하여 내용을 구조화 함. 태그는 ```<>```와 ```</>``` 사이에 본문을 작성하여 문서의 구조를 표현함

## HTML 코드 예시
```
<html>
    <head>
        <title>InterviewBit</title>
    </head>
    <body>
        <h1>My First Web Page<h1>
        <h2>I am heading 2</h2>
        <h3>I am heading 3</h3>
        <!-- Comment: Below is my paragraph -->
        <p>Hey, there I am a paragraph </p>
    </body>
</html>
```

### html 태그는 문서가 시작되는 루트 태그

# BeautifulSoup을 활용하여 해당하는 HTML 정보 가져오기

## Beautifulsoup 이란?
![Alt text](https://lh3.googleusercontent.com/o9HtAcCnpfW_o5b1lkhvrJ0lzZBJ6Lm8TwxYue4Z3K5OdekeptiGVAUEPcBC_1ra7cFqAV0QOFByNl3ub_1BJbNe3A=w640-h400-e365-rj-sc0x00ffffff "bs4")

> *” 뷰티풀 수프(Beautiful Soup)는 HTML과 XML 문서들의 구문을 분석하기 위한 파이썬 패키지이다. HTML로부터 데이터를 추출하기 위해 사용할 수 있는 파싱된 페이지의 파스 트리를 만드는데, 이는 웹 스크래핑에 유용하다. ” (출처: wikipedia)*

## 파이썬의 BeautifulSoup는 XML이나 HTML과 같이 정형화 된 문서에서 필요한 데이터를 추출하기 위해 사용되는 파이썬 라이브러리임. 이를 활용하여 HTML의 **태그** 에 쉽게 접근하고 데이터를 추출할 수 있음

# 우리는 "파싱" 을 통하여 수프에 쌓인 데이터 더미를 정보로 가공할 수 있음

## 파싱 (Parsing)
> *” Parsing, syntax analysis, or syntactic analysis is the process of analyzing a string of symbols, either in natural language, computer languages or data structures, conforming to the rules of a formal grammar. ” (출처: wikipedia)*


## (웹)파싱은 주어진 자연어나 컴퓨터 언어 등과 같이 주어진 일련의 문자열들을 분석하는 프로세스임.

````python
# 필요한 모듈 불러오기
from bs4 import BeautifulSoup # 정형화 된 문서에서 데이를 추출하기 위한 모듈
import requests #웹 HTTP 요청을 위한 모듈
import lxml #정형화 된 문서의 문자열 프로세스 분석을 위한 파서
import pandas as pd
import re
````

# 수프 만들기

## 문서를 해석하려면, 문서를 BeautifulSoup 함수에 문서를 넣어주어야 함

## 예시- *이상한 나라의 엘리스* 일부

````python
exampleHtml = """
<html><head><title>The Dormouse's story</title></head>

<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""

soup = BeautifulSoup(exampleHtml, 'lxml')
soup
````

# soup 함수 활용하여, HTML 형태의 문서를 태그에 따라 데이터를 추출할 수 있음

# 무신사 스탠다드에서 판매되고 있는 의류의 정보를 획득하기 위하여, 무신사 스토어에 마련된 url로부터 문서 획득

````python
url = 'https://www.musinsa.com/brands/musinsastandard?category3DepthCodes=&category2DepthCodes=&category1DepthCode=&colorCodes=&startPrice=&endPrice=&exclusiveYn=&includeSoldOut=&saleGoods=&timeSale=&includeKeywords=&sortCode=emt_high&tags=&page=1&size=90&listViewType=small&campaignCode=&groupSale=&outletGoods=false&boutiqueGoods='
response = requests.get(url)
````

````python
# lxml 파서를 활용하여, 획득한 웹 페이지를 구조화
soup = BeautifulSoup(response.text, 'lxml')
print(soup) # 실제로 생긴 것에 매우 복잡하게 생긴 웹 페이지
````

## 위에서 찾았던 것처럼, 우리가 원하는 정보는 태그 'a' 에서 'img-block' 라는 클래스에 속해있음

````python
soup.find_all('a', attrs={'class':'img-block'})[0:3]
````

````python
product_list = [soup.find_all('a', attrs={'class':'img-block'})[0]['title']]
url_list = soup.find_all('a', attrs={'class':'img-block'})[0]['href']

df = pd.DataFrame({'상품명': product_list,
                  'url': url_list}) 

print(df)
````

## 가격 정보는 태그 'p'에서 'price'라는 클래스에 속해있음

````python
soup.find_all('p', attrs={'class':'price'})[0].get_text() # 정제된 가격을 얻기 위해 text로 값을 획득
````

````python
soup.find_all('p', attrs={'class':'price'})[0:10] #갑자기 시행하는 에누리로 인해 일부 제품은 2개의 가격이 포착된 상황
````

## 정상 가격 상품의 경우 단독 가격을 획득해야 하고, 에누리가 들어간 제품의 경우, 뒤쪽에 위치한 가격을 가져와야 함
## 이럴 경우, if문과 len 함수의 조합으로 값 획득

````python
if len(soup.find_all('p', attrs={'class':'price'})[0].get_text().split()) > 1:
    print(soup.find_all('p', attrs={'class':'price'})[0].get_text().split()[1])
else:
    print(soup.find_all('p', attrs={'class':'price'})[0].get_text().split()[0])
````

## 획득한 금액 정보엔 ',' 와 '원' 이라고 하는 문자열이 붙어있어, 숫자 연산이 어려움. 따라서, 정규표현식을 통해 숫자만 추출이 필요함
## 정규표현식이란? 


>*정규 표현식(正規表現式 / Regular Expression)프로그래밍에서 문자열을 다룰 때, 문자열의 일정한 패턴을 표현하는 일종의 형식 언어를 말한다. 정규식이라고도 부르며, 보통 RegEx 혹은 RegExp라 많이 쓴다. (출처: 나무위키)* 

### "15,900원" 이라고 하는 텍스트에는, 숫자형으로 표현될 수 없는 "," 와 "원" 이라는 텍스트가 포함되어 있기 때문에 아래와 같은 방법을 사용:

````python
re.sub(r'[^0-9]', '', soup.find_all('p', attrs={'class':'price'})[0].get_text().split()[0])
````

````python
start = 1
title_list = []
url_list = []
price_list = []

while start<57: #1페이지~56페이지
    try:
        url = f'https://www.musinsa.com/brands/musinsastandard?category3DepthCodes=&category2DepthCodes=&category1DepthCode=&colorCodes=&startPrice=&endPrice=&exclusiveYn=&includeSoldOut=&saleGoods=&timeSale=&includeKeywords=&sortCode=emt_high&tags=&page={start}&size=90&listViewType=small&campaignCode=&groupSale=&outletGoods=false&boutiqueGoods='.format(start)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'lxml')
        # 상품명과 주소 획득
        for s in soup.find_all('a', attrs={'class':'img-block'}):
            title_list.append(s['title'])
            url_list.append('https:' + s['href'])
        # 상품의 가격을 획득
        for s in soup.find_all('p', attrs={'class':'price'}):
            price = s.get_text().split()
            if len(price) >1:
                price_list.append(re.sub(r'[^0-9]', '', price[1]))
            else:
                price_list.append(re.sub(r'[^0-9]', '', price[0]))
            
        start += 1
    
    except:
        print(start)
        break
````

````python
# 각자 list에 넣은 상품명, URL, 가격을 하나의 dataframe으로 병합
df = pd.DataFrame({'상품명': title_list,
                  'url': url_list,
                 '가격': price_list})
````
